```python
import requests

from bs4 import BeautifulSoup


def get_mp3_urls_from_url(url):
  r = requests.get(url, allow_redirects=True)
  soup = BeautifulSoup(r.content, 'html.parser')
  audio_files = soup.find_all("audio")

  return [audio_file["src"] for audio_file in audio_files]
```

```python
import os

def download_mp3(mp3_url, filename):
  ruta_audios = "/content/drive/MyDrive/audios_peluts"
  ruta_audio = ruta_audios + "/" + filename
  if not os.path.exists(ruta_audio):
    r = requests.get(mp3_url, allow_redirects=True)
    open(ruta_audios + "/" + filename, 'wb').write(r.content)

def download_mp3s(mp3s):
  URL_BASE = "https://www.pelutsproductions.com/"
  mp3s = [URL_BASE + mp3 for mp3 in mp3s]
  for mp3 in mp3s:
    filename = mp3.split("/")[-1]
    download_mp3(mp3, filename)

download_mp3s(mp3s)
```

```python
import torch
from transformers import AutoModelForSpeechSeq2Seq, AutoProcessor, pipeline


device = "cuda:0" if torch.cuda.is_available() else "cpu"
torch_dtype = torch.float16 if torch.cuda.is_available() else torch.float32

model_id = "openai/whisper-large-v3"

model = AutoModelForSpeechSeq2Seq.from_pretrained(
    model_id, torch_dtype=torch_dtype, low_cpu_mem_usage=True, use_safetensors=True
)
model.to(device)

processor = AutoProcessor.from_pretrained(model_id)

pipe = pipeline(
    "automatic-speech-recognition",
    model="openai/whisper-large-v3",
    tokenizer=processor.tokenizer,
    feature_extractor=processor.feature_extractor,
    max_new_tokens=128,
    chunk_length_s=30,
    batch_size=16,
    return_timestamps=True,
    torch_dtype=torch_dtype,
    device=device,
)
```

```python
# Escrapejem tota la web

URL_INICIAL = "https://www.pelutsproductions.com/"

def get_all_urls(url):
  urls = []
  if url not in urls:
    urls.append(url)
    print(url)
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "html.parser")

    for link in soup.find_all("a"):
      href = link.get("href")
      if href is not None and href not in urls and href.startswith("/") or href.startswith(url):
        if href.startswith("/"):
          href = url + href
        urls.append(href)

  return urls

get_all_urls(URL_INICIAL)
```

```python
def get_text_from_url(url):
  r = requests.get(url)
  soup = BeautifulSoup(r.text, "html.parser")
  contents = soup.find_all("div", class_="module-r")
  textos = []
  for content in contents:
      textos.append(content.text)

  return "\n".join(textos)

textos = []
for url in get_all_urls(URL_INICIAL):
  text = get_text_from_url(url)
  textos.append(text)
  print(text)
```

```python
def netejar_textos(textos):
  textos_nets = []

  for text in textos:
    text_sense_espais = text.replace("\t", "").replace('\xa0', "")
    text_en_intros = text_sense_espais.split('\n')
    text_sense_intros = [text for text in text_en_intros if text]
    textos_nets.append("\n".join(text_sense_intros))

  return textos_nets

textos_nets = netejar_textos(textos)

textos_nets
```

```python
with open('/content/drive/MyDrive/audios_peluts/textos_web.txt', 'w', encoding="utf-8") as textos_file:
  for text in textos_nets:
    try:
      textos_file.write(text + '\n')
    except Exception as e:
      print(text)
```
